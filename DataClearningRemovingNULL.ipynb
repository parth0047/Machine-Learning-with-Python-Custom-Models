```
# importing libraries
import numpy as np
import pandas as pd
```

```
dataset = pd.read_csv("social media influencers - youtube.csv")
dataset
```

```
dataset.info()
```

```

# One time Executable code ⚠️💀💀💀💀⚠️

def convert_to_number(x):
  if x is not  np.NaN:
    if x.endswith('M'):
      return float(x[:-1]) * 1_000_000
    elif x.endswith('K'):
      return float(x[:-1]) * 1_000
    else:
      return float(x)
  else:
    pass

dataset['avg views'] = dataset['avg views'].apply(convert_to_number)
dataset['avg likes'] = dataset['avg likes'].apply(convert_to_number)
dataset['avg comments'] = dataset['avg comments'].apply(convert_to_number)
dataset['Subscribers'] = dataset['Subscribers'].apply(convert_to_number)
dataset
```

```
dataset.isnull().sum()
```

```
categories  = dataset['Category'].dropna().unique().tolist()
```

```
country = dataset['Audience Country'].dropna().unique().tolist()
```

```
#Putting Values in string list
import random
def fitRandom(value):
  if pd.isnull(value):
    return random.choice(categories)
  else:
    return(value)
dataset['Category'] = dataset['Category'].apply(fitRandom)
dataset
```

```
#Putting Values in string list
import random
def fitRandom(value):
  if pd.isnull(value):
    return random.choice(country)
  else:
    return(value)
dataset['Audience Country'] = dataset['Audience Country'].apply(fitRandom)
dataset
```

```
dataset.info()
```

```
likes = dataset['avg likes'].dropna().unique().tolist()
```

```
import random
def fitRandom(value):
  if pd.isnull(value):
    return random.choice(likes)
  else:
    return(value)
dataset['avg likes'] = dataset['avg likes'].apply(fitRandom)
dataset
```

```
comments = dataset['avg comments'].dropna().unique().tolist()
```

```
import random
def fitRandom(value):
  if value is np.nan:
    return random.choice(comments);
  else:
    return(value)
dataset['avg comments'] = dataset['avg comments'].apply(fitRandom)
dataset
```
